```python
file_names = 'non-overlapping_includingbidirectional_all_genes_newannotation'
gat_TFBS_constitutive_output_openchrom = f'../../data/output/{file_names}/TFBS_enrichment/gat_analysis/gat_400bp_promoters_5UTR_Czechowski_TFBS_openchrom_constitutive.out'
gat_TFBS_variable_output_openchrom = f'../../data/output/{file_names}/TFBS_enrichment/gat_analysis/gat_400bp_promoters_5UTR_Czechowski_TFBS_openchrom_variable.out'
gat_TFBS_constitutive_output = f'../../data/output/{file_names}/TFBS_enrichment/gat_analysis/gat_400bp_promoters_5UTR_Czechowski_TFBS_constitutive.out'
gat_TFBS_variable_output = f'../../data/output/{file_names}/TFBS_enrichment/gat_analysis/gat_400bp_promoters_5UTR_Czechowski_TFBS_variable.out'
output_prefix = 'Czechowski_400bp_promoters_5UTR'
```

```python
def create_plot(gat_TFBS_constitutive_output,gat_TFBS_variable_output):
    """import and process the raw outputs after running gat (Genomic association tester). Then create barplot of constitutive and variable gene TATA enrichment"""
    #import gat output files as dfs
    constitutive = pd.read_table(gat_TFBS_constitutive_output, sep='\t', header=0)
    variable = pd.read_table(gat_TFBS_variable_output, sep='\t', header=0)
    #filter qvalue under 0.05
    constitutive = constitutive[constitutive.qvalue < 0.05]
    variable = variable[variable.qvalue < 0.05]
    
    return constitutive,variable
    #merge dfs
    merged = pd.concat([constitutive,variable], ignore_index=True)
    #set style to ticks
    sns.set(style="ticks", color_codes=True)
    #bar chart, 95% confidence intervals
    plot = sns.barplot(x="annotation", y="l2fold", data=merged)
    plot.axhline(0, color='black')
    plt.xlabel("Gene type")
    plt.ylabel("Log2-fold enrichment over background").get_figure().savefig(f'../../data/output/{file_names}/TFBS_enrichment/plots/{output_prefix}_log2fold.pdf', format='pdf')
```

```python
#Create barplot
constitutive_openchrom,variable_openchrom = create_plot(gat_TFBS_constitutive_output_openchrom,gat_TFBS_variable_output_openchrom)
constitutive,variable = create_plot(gat_TFBS_constitutive_output,gat_TFBS_variable_output)
```
