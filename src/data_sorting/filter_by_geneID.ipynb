```python
#select_genes_file = '../../data/genomes/ara_housekeeping_list.out'
select_genes_file ='/home/witham/Documents/pipeline_new/PromoterArchitecture/data/output/non-overlapping_includingbidirectional_all_genes_newannotation_2KB/W5_gene_list.txt'

all_promoters_bedfile = '/home/witham/Documents/pipeline_new/PromoterArchitecture/data/output/non-overlapping_includingbidirectional_all_genes_newannotation_2KB/FIMO/promoters.bed'
#'../../data/FIMO/promoters_renamedChr.bed'
```

```python
select_genes = pd.read_table(select_genes_file, sep='\t', header=None)
cols = ['AGI']#,'gene_type'
select_genes.columns = cols
all_promoters = pd.read_table(all_promoters_bedfile, sep='\t', header=None)
cols = ['chr', 'start', 'stop', 'AGI', 'dot', 'strand', 'source', 'type', 'dot2', 'details']
all_promoters.columns = cols
```

### filter out unwanted genes

```python
merged = pd.merge(select_genes, all_promoters, on='AGI')
```

### change column order to bed file format

```python
filtered_proms = merged.loc[:, ['chr', 'start', 'stop', 'AGI', 'dot', 'strand', 'source', 'type', 'dot2', 'details']]
#sort the df by chromosome then by motif start position
filtered_proms_sorted = filtered_proms.sort_values(['chr','start'])
```
