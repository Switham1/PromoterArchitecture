```python
import pandas as pd
```

```python
echo = pd.read_csv('echoorder.csv', header=0)
echo
```

```python
guides_map = pd.read_csv('lvl1guides.csv', header=0)
guides_map.rename(columns={'DNA stock to add to reaction (nL) echo': 'Vol_nL'}, inplace=True)
guides_map
```

```python
merged = pd.merge(echo,guides_map,how='left',left_on='Module3ID', right_on='Plasmid')
merged = merged[['Module3ID','Vol_nL','Module4ID',]]
merged = pd.merge(merged,guides_map,how='left',left_on='Module4ID', right_on='Plasmid')
merged = merged[['Module3ID','Vol_nL_x','Module4ID','Vol_nL_y']]
merged.to_csv('merged.csv',header=1, index=False)
```
